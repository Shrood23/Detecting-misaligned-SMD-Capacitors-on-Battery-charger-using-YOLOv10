```python
import os
from ultralytics import YOLO
import cv2

def calculate_iou(box1, box2):
    """Tính chỉ số IoU giữa hai bounding box."""
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    # Diện tích phần giao nhau
    intersection = max(0, x2 - x1) * max(0, y2 - y1)

    # Diện tích mỗi box
    area_box1 = (box1[2] - box1[0]) * (box1[3] - box1[1])
    area_box2 = (box2[2] - box2[0]) * (box2[3] - box2[1])

    # Diện tích phần hợp
    union = area_box1 + area_box2 - intersection

    # Trả về IoU
    return intersection / union if union > 0 else 0

# Tải mô hình
model = YOLO("C:\\Do_an\\weights\\run3\\weights\\best.pt")

# Load ảnh
image_path = "C:\\Do_an\\dataset\\images\\test\\bad (449).jpg"
img = cv2.imread(image_path)

# Chạy mô hình
results = model(img)

# Lấy danh sách box và thông tin
bad_detected = False
low_confidence_detected = False
overlapping_boxes_detected = False

# Lấy thông tin các bounding box
boxes = []
for box in results[0].boxes:
    label = int(box.cls[0].item())  # Lấy nhãn (index)
    conf = box.conf[0].item()  # Lấy độ tin cậy
    class_name = model.names[label]  # Lấy tên class từ model

    # Thêm thông tin box vào danh sách
    x1, y1, x2, y2 = box.xyxy[0].tolist()  # Lấy tọa độ box
    boxes.append((x1, y1, x2, y2, class_name, conf))

    # Kiểm tra điều kiện
    if class_name == "bad":  # Nếu có nhãn "bad"
        bad_detected = True
    elif class_name == "good" and conf < 0.2:  # Nếu nhãn "good" có độ tin cậy thấp
        low_confidence_detected = True

# Kiểm tra các box có trùng nhau không
for i in range(len(boxes)):
    for j in range(i + 1, len(boxes)):
        box1 = boxes[i][:4]
        box2 = boxes[j][:4]
        iou = calculate_iou(box1, box2)
        if iou > 0.5:  # Ngưỡng để coi là trùng nhau
            overlapping_boxes_detected = True
            break
    if overlapping_boxes_detected:
        break

# Quyết định mạch là "bad" hay "good"
if bad_detected:
    circuit_status = "bad"
elif low_confidence_detected:
    circuit_status = "bad"
elif overlapping_boxes_detected:
    circuit_status = "bad"
else:
    circuit_status = "good"

# Lưu ảnh đã nhận diện vào folder khác
output_folder = "C:\\Do_an\\detected_images"  # Đường dẫn folder đích
os.makedirs(output_folder, exist_ok=True)  # Tạo folder nếu chưa tồn tại

output_path = os.path.join(output_folder, os.path.basename(image_path))  # Đường dẫn file
annotated_image = results[0].plot()  # Ảnh đã vẽ bounding box
cv2.imwrite(output_path, annotated_image)  # Lưu ảnh

print(f"Trạng thái mạch: {circuit_status}")
print(f"Ảnh đã lưu tại: {output_path}")
```
