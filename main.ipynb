```python
import os
from ultralytics import YOLO
import cv2

def calculate_iou(box1, box2):
    """TÃ­nh chá»‰ sá»‘ IoU giá»¯a hai bounding box."""
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    # Diá»‡n tÃ­ch pháº§n giao nhau
    intersection = max(0, x2 - x1) * max(0, y2 - y1)

    # Diá»‡n tÃ­ch má»—i box
    area_box1 = (box1[2] - box1[0]) * (box1[3] - box1[1])
    area_box2 = (box2[2] - box2[0]) * (box2[3] - box2[1])

    # Diá»‡n tÃ­ch pháº§n há»£p
    union = area_box1 + area_box2 - intersection

    # Tráº£ vá» IoU
    return intersection / union if union > 0 else 0

# Táº£i mÃ´ hÃ¬nh
model = YOLO("C:\\Do_an\\weights\\run3\\weights\\best.pt")

# Load áº£nh
image_path = "C:\\Do_an\\dataset\\images\\test\\bad (449).jpg"
img = cv2.imread(image_path)

# Cháº¡y mÃ´ hÃ¬nh
results = model(img)

# Láº¥y danh sÃ¡ch box vÃ  thÃ´ng tin
bad_detected = False
low_confidence_detected = False
overlapping_boxes_detected = False

# Láº¥y thÃ´ng tin cÃ¡c bounding box
boxes = []
for box in results[0].boxes:
    label = int(box.cls[0].item())  # Láº¥y nhÃ£n (index)
    conf = box.conf[0].item()  # Láº¥y Ä‘á»™ tin cáº­y
    class_name = model.names[label]  # Láº¥y tÃªn class tá»« model

    # ThÃªm thÃ´ng tin box vÃ o danh sÃ¡ch
    x1, y1, x2, y2 = box.xyxy[0].tolist()  # Láº¥y tá»a Ä‘á»™ box
    boxes.append((x1, y1, x2, y2, class_name, conf))

    # Kiá»ƒm tra Ä‘iá»u kiá»‡n
    if class_name == "bad":  # Náº¿u cÃ³ nhÃ£n "bad"
        bad_detected = True
    elif class_name == "good" and conf < 0.2:  # Náº¿u nhÃ£n "good" cÃ³ Ä‘á»™ tin cáº­y tháº¥p
        low_confidence_detected = True

# Kiá»ƒm tra cÃ¡c box cÃ³ trÃ¹ng nhau khÃ´ng
for i in range(len(boxes)):
    for j in range(i + 1, len(boxes)):
        box1 = boxes[i][:4]
        box2 = boxes[j][:4]
        iou = calculate_iou(box1, box2)
        if iou > 0.5:  # NgÆ°á»¡ng Ä‘á»ƒ coi lÃ  trÃ¹ng nhau
            overlapping_boxes_detected = True
            break
    if overlapping_boxes_detected:
        break

# Quyáº¿t Ä‘á»‹nh máº¡ch lÃ  "bad" hay "good"
if bad_detected:
    circuit_status = "bad"
elif low_confidence_detected:
    circuit_status = "bad"
elif overlapping_boxes_detected:
    circuit_status = "bad"
else:
    circuit_status = "good"

# LÆ°u áº£nh Ä‘Ã£ nháº­n diá»‡n vÃ o folder khÃ¡c
output_folder = "C:\\Do_an\\detected_images"  # ÄÆ°á»ng dáº«n folder Ä‘Ã­ch
os.makedirs(output_folder, exist_ok=True)  # Táº¡o folder náº¿u chÆ°a tá»“n táº¡i

output_path = os.path.join(output_folder, os.path.basename(image_path))  # ÄÆ°á»ng dáº«n file
annotated_image = results[0].plot()  # áº¢nh Ä‘Ã£ váº½ bounding box
cv2.imwrite(output_path, annotated_image)  # LÆ°u áº£nh

print(f"Tráº¡ng thÃ¡i máº¡ch: {circuit_status}")
print(f"áº¢nh Ä‘Ã£ lÆ°u táº¡i: {output_path}")
```
